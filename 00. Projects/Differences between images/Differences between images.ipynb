```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

Load images, that we will compare in varables

```python
img1 = Image.open('images/image_1.jpg')
img2 = Image.open('images/image_2.jpg')
```

```python
fig, axs = plt.subplots(1, 2)

axs[0].imshow(img1, cmap='gray')
axs[1].imshow(img2, cmap='gray')

plt.show()
```

```python
img1 = img1.convert('L')
img2 = img2.convert('L')
```

```python
fig, axs = plt.subplots(1, 2)

axs[0].imshow(img1, cmap='gray')
axs[1].imshow(img2, cmap='gray')

plt.show()
```

```python
img1_npa = np.array(img1)
img2_npa = np.array(img2)
diff = img1_npa - img2_npa
```

```python
plt.imshow(diff, cmap='gray')
plt.show()
```

```python
diff
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(np.invert(diff), cmap='gray')
plt.show()
```

```python
diff_abs = np.abs(img1_npa - img2_npa)

#      
threshold = 200

#  ,       -  
diff_mask = np.where(diff > threshold, 0, diff_abs)

#  
plt.figure(figsize=(8, 8))
plt.imshow(np.invert(diff_mask), cmap='gray')
plt.show()
```
